```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harshshinde8/movies-csv")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
df=pd.read_csv(f"{path}/movies.csv")
#preprocessing step
df["genres"]=df["genres"].fillna("")
df["overview"]=df["overview"].fillna("")
df["content"]=df["genres"]+" "+df['overview']
#TF-IDF Vectorizer
tfidf=TfidfVectorizer(stop_words="english")
tfidf_matrix=tfidf.fit_transform(df["content"])
cosine_sim=cosine_similarity(tfidf_matrix,tfidf_matrix)
#droping the duplicates
indices=pd.Series(df.index,index=df["original_title"]).drop_duplicates()
def recommendation_movie(title,topn=5):
  if title not in indices:
    return "Movie is not found!"
  idx=indices[title]
  sim_scores=list(enumerate(cosine_sim[idx]))
  sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
  sim_scores=sim_scores[1:topn+1]
  movie_indices=[i[0] for i in sim_scores]
  return df["original_title"].iloc[movie_indices]
recommendation_movie("Inception")
```

### Recommend Top-10 Series from IMDB Dataset

```python
import pandas as pd

# Load the IMDB dataset
imdb_df = pd.read_csv('/content/imdb_top_1000.csv')

# Display the first few rows and column
display(imdb_df.head())
```

```python
# Recommendation

series_df = imdb_df[imdb_df['Genre'].str.contains('TV Series|Documentary|Animation', na=False, regex=True)]
print(f"Number of TV Series found: {len(series_df)}")

# Sort by 'IMDB_Rating' in descending order and then by 'No_of_Votes' to break ties
top_10_series = series_df.sort_values(by=['IMDB_Rating', 'No_of_Votes'], ascending=[False, False]).head(10)

print("\nTop 10 Recommended Series:")
display(top_10_series[['Series_Title', 'IMDB_Rating', 'No_of_Votes', 'Genre', 'Director', 'Star1', 'Star2', 'Star3', 'Star4']].reset_index(drop=True))
```

### Observations

The top 10 recommended series are listed above, sorted primarily by 'IMDB_Rating' and then by 'No_of_Votes'.

*   **High Ratings:** All the recommended series have exceptionally high IMDB ratings, typically 8.0 and above, indicating their critical acclaim and popularity among viewers.
*   **Diverse Genres:** The list includes a variety of genres such as Animation, Documentary, and Action, suggesting that highly-rated series are not confined to a single genre.
*   **Popularity:** The 'No_of_Votes' column provides an insight into how many people have rated these series, reinforcing their popularity.
*   **Filtering:** The filtering logic for 'TV Series' was based on the 'Genre' column. It's important to note that the dataset might not explicitly categorize all 'series' in a straightforward manner, and some might be long-running documentaries or animated shows. If the goal was strictly live-action TV series, the filtering might need to be refined further (e.g., by checking 'Runtime' or other indicative columns if available).
