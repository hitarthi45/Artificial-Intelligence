```python
#3 sentence toward country, 2 sentence towards sports, 2 sentence towards technology
Sentences =[ " India is my country. All indians are my brothers and sisters",
             " India is country of herbs and spices.",
             " India is known for its rich and varied heritage.",

             " I like to see cricket and football.",
             " Cricket has many talented cricketers like Virat kohli and Rohit sharma.",

             " Now-a-days the technology is growing very fast.",
             " AI industry is taking place in the  market rapidly." ]
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #TF-bow
from sklearn.feature_extraction.text import TfidfVectorizer #TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#embeddings using TF-bow
tf_vectorizer = CountVectorizer()
tf_matrix=tf_vectorizer.fit_transform(Sentences)
```

```python
tf_matrix
```

```python
tf_similarity=cosine_similarity(tf_matrix)
tf_similarity
```

```python
#embeddings using TF-idf
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix=tfidf_vectorizer.fit_transform(Sentences)
tfidf_matrix
```

```python
tfidf_similarity=cosine_similarity(tfidf_matrix)
tfidf_similarity
```

```python
#finding document similarity
def doc_sim(similarity_matrix,Sentences):
  for i in range(len(Sentences)):
    sims=similarity_matrix[i].copy()
    sims[i]=-1
    top_index=np.argmax(sims)
    print(Sentences[i],"-->",Sentences[top_index])
```

```python
doc_sim(tfidf_similarity,Sentences)
```

### POST LAB


```python
documents = [
    Sentences[0], # "India is my country. All indians are my brothers and sisters"
    Sentences[1], # "India is country of herbs and spices."
    Sentences[2]  # "India is known for its rich and varied heritage."
]

print("Documents selected:")
for i, doc in enumerate(documents):
    print(f"Document {i+1}: {doc}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# BoW Vectorization
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(documents)

print("BoW Matrix (Term Frequencies):")
print(bow_matrix.toarray())
print("\nFeatures (Vocabulary):")
print(bow_vectorizer.get_feature_names_out())

# Calculate Cosine Similarity for BoW
bow_similarity = cosine_similarity(bow_matrix)

print("\nBoW Similarity Matrix:")
print(bow_similarity)
```

```python
# TF Vectorization (same as BoW with CountVectorizer)
tf_vectorizer = CountVectorizer()
tf_matrix = tf_vectorizer.fit_transform(documents)

print("TF Matrix (Term Frequencies):")
print(tf_matrix.toarray())

# Calculate Cosine Similarity for TF
tf_similarity = cosine_similarity(tf_matrix)

print("\nTF Similarity Matrix:")
print(tf_similarity)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
print("\nFeatures (Vocabulary):")
print(tfidf_vectorizer.get_feature_names_out())

# Calculate Cosine Similarity for TF-IDF
tfidf_similarity = cosine_similarity(tfidf_matrix)

print("\nTF-IDF Similarity Matrix:")
print(tfidf_similarity)
```
