```python
text="""I am Hitarthi Pansuriya learning the AI subject in MA104 taught by Professor Nishith Kotak and he is teaching NLP basics to sem-6 students.
 His lecture is of 1 and haf hour. He is teaching this subject for 4 years in Marwadi university in ICT department."""
```

```python
import spacy
nlp1=spacy.load('en_core_web_sm')
doc=nlp1(text)
```

```python
for token in doc.ents:
  print(token.text, token.label_)
```

```python
#POS : Part of Speech
text = """Let's google it """
doc = nlp1(text)
for token in doc:
  print(token.text, "-->", token.pos_)
```

```python
#stemming and Lemmatixation
text="""I studied AI subject and
then meeting Mr. Sachin in a meeting"""
```

```python
doc=nlp1(text)
doc
```

```python
#stemming
import nltk
from nltk.stem.porter import *

p_stem=PorterStemmer()
for word in text.split():
  print(word,"-->",p_stem.stem(word))
```

```python
#lemmatization
for token in nlp(text):
  print(token.text,"-> ",token.pos_,
        "->",token.lemma_)
```
