```python
import nltk
from nltk.util import ngrams
from collections import Counter

try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')
```

```python
def generate_ngrams(text, n):
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text.lower())

    # Generate n-grams
    n_grams = list(ngrams(tokens, n))

    # Count the frequency
    ngram_counts = Counter(n_grams)

    return ngram_counts


text = """As the sun dipped below the horizon, painting the sky in vibrant shades of orange and purple, the old lighthouse began its rhythmic, sweeping rotation,
casting a beam of comforting light across the treacherous, crashing waves to guide the weary sailors toward the safety of the harbor."""

#bigrams (N=2)
bigrams = generate_ngrams(text, 2)
print("\nBigrams (N=2):")
for ngram, count in bigrams.most_common(5):
    print(f"'{' '.join(ngram)}': {count}")

#trigrams (N=3)
trigrams = generate_ngrams(text, 3)
print("\nTrigrams (N=3):")
for ngram, count in trigrams.most_common(5):
    print(f"'{' '.join(ngram)}': {count}")

#4-grams (N=4)
four_grams = generate_ngrams(text, 4)
print("\n4-grams (N=4):")
for ngram, count in four_grams.most_common(5):
    print(f"'{' '.join(ngram)}': {count}")
```

```python
import nltk
from nltk.util import ngrams
from collections import Counter

def generate_ngrams(text, n):
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text.lower())  # Convert to lowercase for consistent n-grams

    # Generate n-grams
    n_grams = list(ngrams(tokens, n))

    # Count the frequency of each n-gram
    ngram_counts = Counter(n_grams)

    return ngram_counts

def predict_next_word(seed_phrase, n_gram_counts, n):
    # Convert seed phrase to lowercase tokens
    seed_tokens = nltk.word_tokenize(seed_phrase.lower())

    if len(seed_tokens) != n - 1:
        print(f"Error: Seed phrase length ({len(seed_tokens)}) must be n-1 ({n-1}) for {n}-gram prediction.")
        return None

    possible_next_words_counts = Counter()

    for ngram, count in n_gram_counts.items():
        if ngram[:-1] == tuple(seed_tokens): # Check if the ngram starts with the seed phrase
            possible_next_words_counts[ngram[-1]] += count

    if not possible_next_words_counts:
        return "(No prediction: Seed phrase not found in n-grams)"

    # Get the most common next word
    most_common_word = possible_next_words_counts.most_common(1)[0][0]

    return most_common_word

# Sample document
text = """As the sun dipped below the horizon, painting the sky in vibrant shades of orange and purple, the old lighthouse began its rhythmic, sweeping rotation,
casting a beam of comforting light across the treacherous, crashing waves to guide the weary sailors toward the safety of the harbor. The harbor was safe and calm."""

# For simplicity, let's use trigrams (n=3) for next word prediction
n_for_prediction = 3
trigrams = generate_ngrams(text, n_for_prediction)

# Example 1: Predict next word for "the sun"
seed_phrase_1 = "the sun"
next_word_1 = predict_next_word(seed_phrase_1, trigrams, n_for_prediction)
print(f"\nPredicting next word for '{seed_phrase_1}' (using {n_for_prediction}-grams): {next_word_1}")

# Example 2: Predict next word for "of the" (corrected from "safety of the")
seed_phrase_2 = "of the"
next_word_2 = predict_next_word(seed_phrase_2, trigrams, n_for_prediction)
print(f"\nPredicting next word for '{seed_phrase_2}' (using {n_for_prediction}-grams): {next_word_2}")

# Example 3: Predict next word for a phrase not in the text
seed_phrase_3 = "big red"
next_word_3 = predict_next_word(seed_phrase_3, trigrams, n_for_prediction)
print(f"\nPredicting next word for '{seed_phrase_3}' (using {n_for_prediction}-grams): {next_word_3}")
```
