```python
sentences = """
     I am Hitarthi Pansuriya and I am 20 years old.
     I am Pursuing B.Tech. and currently in 6th semester in ICT engineering.
     I will be graduated in 2027.
     I want to do a web developer job in good software company.
     In future I want to become a successful businesswoman after gaining experience.
"""
```

```python
import nltk
import networkx as nx
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
#dowload requirement
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def textrank_keyword(text, num_keywords=5, window_size=4):
  words = word_tokenize(text.lower())
  stop_words = set(stopwords.words('english'))
  words = [w for w in words if w not in stop_words and w not in string.punctuation]

  graph = nx.Graph()

  # Add nodes to the graph (all unique words)
  for word in set(words):
    graph.add_node(word)

  # Add edges based on window_size
  for i, word in enumerate(words):
    for j in range(i+1, window_size + 1): # k is the offset from the current word
      if j < len(words): # Ensure the connected word is within bounds
        graph.add_edge(word, words[j])

  # Handle case where graph might be empty or disconnected
  if not graph.nodes():
      return []

  scores = nx.pagerank(graph)
  ranked_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
  keywords = [word for word, score in ranked_words[:num_keywords]] # Use num_keywords instead of topn
  return keywords
```

```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
import nltk
from nltk.util import ngrams
from collections import defaultdict,Counter
import random
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens=nltk.word_tokenize(corpus)
```

```python
#build ngram model
n=2
bigrams=list(ngrams(tokens,n))
print(bigrams)
```

```python
#create the probability model
model = defaultdict(Counter)
for w1,w2 in bigrams:
  model[w1][w2]+=1

model
```

```python
#function to predict next word
def predict_nextword(word):
  word = word.lower()
  if word in model:
    next_word = model[word]
    prediction = max(next_word, key=next_word.get)
    return prediction

  else:
    return "No prediction is done"
```

```python
predict_nextword("deep") #-> "learning"
# predict_nextword("I") -> "like"
# predict_nextword("i") -> "no prediction"
# predict_nextword("machine") -> "learnig"
```
